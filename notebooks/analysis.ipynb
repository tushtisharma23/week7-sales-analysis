This notebook performs sales analysis using pandas.

```python
import pandas as pd

df = pd.read_csv("../data/raw/sales_data.csv")
```

```python
df.groupby("category")["total_amount"].sum()
```

```python
df.groupby("product_name")["total_amount"].sum().sort_values(ascending=False).head()
```

```python
df["order_date"] = pd.to_datetime(df["order_date"])

df.groupby(df["order_date"].dt.to_period("M"))["total_amount"].sum()
```
